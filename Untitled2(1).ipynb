```python
image_size = (100, 100)
from keras.preprocessing.image import img_to_array , load_img
def load_images(path):
    images, labels = [], []
    for label in ['cat', 'dog']:
        folder = path + label + 's/'
        for filename in os.listdir(folder):
            filepath = folder + filename
            photo = img_to_array(load_img(filepath, target_size=image_size))
            photo /= 255
            images += [photo]
            labels += [0 if label == 'cat' else 1]
    return np.array(images), np.array(labels)

training_images, training_labels = load_images('training_set/')
test_images, test_labels = load_images('test_set/')
```

***randomize arrays***(Shuffle two list at once with same order)

```python
indices = np.arange(test_images.shape[0])
np.random.shuffle(indices)
X_test = test_images[indices]
Y_test = test_labels[indices]

print(f'type Y_test : {type(Y_test)} , type X_test : {type(X_test)} ')

indices1 = np.arange(training_images.shape[0])
np.random.shuffle(indices1)
X_train  = training_images[indices1]
Y_train  = training_labels[indices1]
```

```python
from keras.utils import np_utils

training_labels = np_utils.to_categorical(training_labels , 2 )
test_labels     = np_utils.to_categorical(test_labels     , 2 )

Y_train = np_utils.to_categorical(Y_train , 2 )
Y_test  = np_utils.to_categorical(Y_test  , 2 )
```

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout , Activation , Flatten
from keras.optimizers import Adam
from keras.losses import categorical_crossentropy

```

```python
my_model = Sequential()
my_model.add(Flatten(input_shape = (100 ,100 , 3 )))
my_model.add(Dense(512 , activation="relu"))
my_model.add(Dense(16  , activation ="relu"))
my_model.add(Dense(2   , activation="softmax"))
my_model.compile(loss = categorical_crossentropy , optimizer = Adam(0.001) , metrics=['accuracy'])
```

we can  also create model with  training_images , training_labels , also for test

```python
history = my_model.fit(X_train, Y_train, epochs = 100, batch_size=32 , verbose = 2 , validation_data = (X_test , Y_test)  )
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss']) 
plt.title('Model loss') 
plt.ylabel('Loss') 
plt.xlabel('Epoch') 
plt.legend(['Train', 'Test'], loc='upper left') 
plt.show()
```
